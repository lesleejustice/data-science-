```python
#recode genre1
def genre (series):
  if series == "genre1":
    return "0"
  if series == "action": 
    return "1"
  if series == "animation": 
    return "2"
  if series == "comedy":
    return "3"
  if series == "crime":
    return "4"
  if series == "documentation":
    return "5"
  if series == "drama":
    return "6"
  if series == "family":
    return "7"
  if series == "fantasy":
    return "8"
  if series == "history":
    return "9"
  if series == "horror":
    return "10"
  if series == "music":
    return "11"
  if series == "reality":
    return "12"
  if series == "romance":
    return "13"
  if series == "scifi":
    return "14"
  if series == "sport":
    return "15"
  if series == "thriller":
    return "16"
  if series == "war":
    return "17"
  if series == "western":
    return "18"
streamingdata["genre"] = streamingdata["genre1"].apply(genre)
```

```python
streamingdata.head()
```

```python
#recode age_certification
def age (series):
  if series == "age_certification":
    return "0"
  if series == "G": 
    return "1"
  if series == "NC-17": 
    return "2"
  if series == "PG":
    return "3"
  if series == "PG-13":
    return "4"
  if series == "R":
    return "5"
  if series == "TV-14":
    return "6"
  if series == "TV-G":
    return "7"
  if series == "TV-MA":
    return "8"
  if series == "TV-PG":
    return "9"
  if series == "TV-Y":
    return "10"
  if series == "TV-Y7":
    return "11"
streamingdata["age"] = streamingdata["age_certification"].apply(age)
```

```python
#recode type
def program_type (series):
  if series == "type":
    return "0"
  if series == "SHOW": 
    return "1"
  if series == "MOVIE": 
    return "2"
streamingdata["program_type"] = streamingdata["type"].apply(program_type)
```

```python
def decade (series):
  if series == "release_year":
    return "0"
  if series <= "1969": 
    return "1"
  if series <= "1979": 
    return "2"
  if series <= "1989":
    return "3"
  if series <= "1999":
    return "4"
  if series <= "2009":
    return "5"
  if series <= "2019":
    return "6"
  if series <= "2029":
    return "7"
streamingdata["decade"] = streamingdata["release_year"].apply(decade)
```
